## Q1. Explain GET and POST methods.

- Get --->  The GET method is to request data from the server. It's used when the client wants to retrieve information from the server, such as a webpage, a resource, or data.

- Caching: GET requests can be cached by browsers and intermediate systems (proxies, CDNs) because they are considered safe and idempotent. This means if the same GET request is made multiple times, the result should be the same, and it should not have any side effects on the server.

- Security: It's less secure for sensitive data because the data is visible in the URL and can be easily bookmarked or shared.

- POST ---> The POST method is to send data to the server for processing. It's used when the client wants to submit data, such as form inputs, to the server, which may result in the server storing or updating data.

- Caching: POST requests are not typically cached by browsers or intermediate systems because they can have side effects on the server, making them non-idempotent.

- Security: More secure for sensitive data because the data is not directly visible in the URL. However, additional security measures (e.g., encryption, authentication) are still necessary for truly sensitive information.

# Q2. Why is request used in Flask?

- In Flask, the request object is used to handle incoming HTTP requests from clients (usually web browsers) to the server. It allows you to access various aspects of the incoming request, such as query parameters, form data, headers, cookies, and more. The request object is a crucial part of building web applications in Flask because it enables you to interact with and process data sent by the client.

## Q4. What are templates in Flask? Why is the render_template() function used?

- Templates are used to separate the presentation logic from the application's code. They provide a way to generate dynamic HTML content by embedding variables, conditional statements, loops, and other logic within the HTML structure. Templates allow you to create consistent and reusable layouts for your web pages, making it easier to manage the design and structure of your application.

- The render_template() function in Flask is used to render these templates, filling them with dynamic data and generating HTML content to be sent to the client's web browser. This function is essential for building dynamic web pages that can display data from the application's backend.
